### 1. DFS with Stack 

```python
def dfs_stack(graph, start):
    visited = set()
    stack = [start] 
    while stack:
        vertex = stack.pop()        
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            for neighbor in reversed(graph[vertex]):
                if neighbor not in visited:
                    stack.append(neighbor)


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

dfs_stack(graph,'A')
```

### 2. Research about "Inorder, Preorder, Postorder" and implement in DFS

```python
structure = {
    'A': {'left': 'B', 'right': 'C'},
    'B': {'left': 'D', 'right': 'E'},
    'C': {'left': None, 'right': 'F'},
    'D': {'left': None, 'right': None},
    'E': {'left': None, 'right': None},
    'F': {'left': None, 'right': None}
}


def traverse_pre(structure, root):
    if root:
        print(root, end=" ")
        traverse_pre(structure, structure[root]['left'])
        traverse_pre(structure, structure[root]['right'])


def traverse_in(structure, root):
    if root:
        traverse_in(structure, structure[root]['left'])
        print(root, end=" ")
        traverse_in(structure, structure[root]['right'])


def traverse_post(structure, root):
    if root:
        traverse_post(structure, structure[root]['left'])
        traverse_post(structure, structure[root]['right'])
        print(root, end=" ")


print("Preorder DFS:")
traverse_pre(structure, 'A')

print("\nInorder DFS:")
traverse_in(structure, 'A')

print("\nPostorder DFS:")
traverse_post(structure, 'A')
```
