### 1. BFS without Queue

```python
def bfs_without_queues(graph, start):
    visited = set()
    container = [start] 
    while container:
        vertex = container[0]
        container = container[1:] 
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)

            for neighbor in graph[vertex]:
                if neighbor not in visited:
                    container.append(neighbor)

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

bfs_without_queues(graph,'A')
```

### 2.BFS with Queue

```python
def bfs_with_queues(graph, start):
    visited = set()
    queue = [start]  
    while queue:
        vertex = queue.pop(0)  
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)

            for neighbor in graph[vertex]:
                if neighbor not in visited:
                    queue.append(neighbor)


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

bfs_with_queues(graph,'A')


```
